```python
import os
import time
import requests
from bs4 import BeautifulSoup
from PIL import Image
from io import BytesIO
from IPython.display import display # 코랩에서 이미지를 보여주기 위한 함수

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service # Import Service
from webdriver_manager.chrome import ChromeDriverManager # Import ChromeDriverManager

# 1. 셀레늄 설정 (Colab 필수 설정)
chrome_options = Options()
chrome_options.add_argument('--headless') # 브라우저 창을 띄우지 않음
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-gpu') # Add this line to prevent abnormal exit in headless environments
chrome_options.binary_location = '/usr/bin/google-chrome' # Explicitly set Chrome binary location to google-chrome

# 코랩에 업로드한 파일 경로 (예: /content/rendered_parrot.html)
# 파일을 코랩 왼쪽 폴더 모양 아이콘을 눌러 업로드한 뒤 경로를 수정하세요.
file_path = "file:///content/rendered_parrot.html" # Corrected file path for Colab

# --- (중략: 1. BeautifulSoup 부분은 동일하므로 생략 가능) ---

print("\n=== 2 & 3. Selenium 동적 크롤링 및 이미지 출력 (Colab 버전) ===")
# Use ChromeDriverManager to automatically get the correct chromedriver path
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options) # Use the service object
driver.get(file_path)

try:
    # 1. 버튼 클릭 (JavaScript 실행)
    button = driver.find_element(By.TAG_NAME, "button")
    button.click()
    time.sleep(1) # 이미지 생성 대기

    # 2. 생성된 이미지 요소 찾기 및 HTML 저장
    img_element = driver.find_element(By.TAG_NAME, "img")
    copy_parrot = img_element.get_attribute('outerHTML')
    print(f"변수에 저장된 HTML 태그: {copy_parrot}")

    # 3. 이미지 시각화
    img_url = img_element.get_attribute("src")

    if img_url.startswith("http"):
        response = requests.get(img_url)
        img_display = Image.open(BytesIO(response.content))
    else:
        # 로컬 파일 경로 처리 (코랩 세션에 파일이 있어야 함)
        # src가 'parrot.jpg' 같은 상대 경로일 경우를 대비
        local_img_path = img_url.replace("file://", "")
        if os.path.exists(local_img_path):
            img_display = Image.open(local_img_path)
        else:
            print(f"이미지 파일을 찾을 수 없습니다: {local_img_path}")
            img_display = None

    if img_display:
        print("이미지를 화면에 출력합니다:")
        display(img_display) # 코랩 전용 출력 방식

finally:
    driver.quit()
```
